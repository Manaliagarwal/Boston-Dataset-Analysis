```python
# boston data set

# multivariable regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from sklearn.datasets import make_regression,load_boston
from random import randint
from sklearn.metrics import accuracy_score
from mpl_toolkits.mplot3d import Axes3D
warnings.filterwarnings("ignore")
plt.style.use("ggplot")
%matplotlib notebook
```

```python
# generating linear data

#x,y=make_regression(100,noise=20,n_features=2)
#y=y.reshape(-1,1)

data=load_boston()
x=data.data
y=data.target.reshape(-1,1)
print(x)
print(y)
```

```python
# plotting the data
#k=black color

%matplotlib notebook
fig=plt.figure(figsize=(10,7),dpi=100)
ax=fig.add_subplot(111,projection='3d')
ax.set_facecolor("white")
ax.scatter(x[:,2],x[:,7],y,c="k")
plt.subplots_adjust(top=1.1, bottom=-0.1,right=1,left=-0.2,hspace=0,wspace=0)
```

```python
#linear regression with gradient descent

def calculate_weight(x,y,lr,epoch,print_interval):
    print("Iteration","Error","Bias","\Weight",sep="\t\t\t")
    x=np.insert(x,0,1,axis=1)
    theta=np.zeros(len(x[0]))
    m=y.size
    old_error,error=0,0
    for i in range(epoch):
        h=(x@theta).reshape(-1,1)
        error=((y-h)**2).mean()
        theta=theta-lr*(1/m)*(x.T.dot(h-y)).reshape(-1)
        print(i+1,np.round(error,2),"",*np.round(theta,3)[:6],sep="\t") if not i%print_interval else""
        if abs(old_error-error)<0.001:
            break
        old_error=error
    print(i+1,np.round(error,2),"",*np.round(theta,3)[:6],sep="\t")
    return theta

def Predict(p,theta):
    x=np.insert(p,0,1,axis=0)
    return x.dot(theta)

    
```

```python
theta=calculate_weight(x,y,0.000001,20000,1000)
```

```python
y_cap=np.array([Predict(i,theta) for i in x])
fig=plt.figure(figsize=(10,7),dpi=100)
ax=fig.add_subplot(111,projection='3d')
ax.set_facecolor("white")
ax.scatter(x[:,2],x[:,7],y,c="r")
ax.scatter(x[:,2],x[:,7],y_cap,c="aqua")
plt.subplots_adjust(top=1.1,bottom=-0.1,right=1,left=-0.2,hspace=0,wspace=0)
```
